```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
nltk.download('punkt')   
from nltk import word_tokenize, sent_tokenize
from nltk.stem.snowball import SnowballStemmer
```

```python
def normalize(text):
  return text.lower()
```

```python
def remove_stopwords(text):
  list_stopwords = stopwords.words("english")
  final_text = ' '.join(a for a in word_tokenize(text) if (a not in list_stopwords and a.isalnum()))
  return final_text
```

```python
def remove_numbers(text):
  text = re.sub("\d+", "",text)
  return text
```

```python
def remove_spaces(text):
  text = re.sub("\s{2,}", "",text)
  return text
```

```python
def sent_tok(text):
  s = re.sub("\n", " ",text)
  s = sent_tokenize(s)
  temp= []
  for sent in s:
    temp.append(re.sub("\W", " ",sent))
  return temp
```

```python
def stem_text(text):
  stemmer = SnowballStemmer("english")
  t  = ' '.join(stemmer.stem(a) for a in word_tokenize(text))
  return t
```

```python
def preprocess(text):
  text = normalize(text)
  text = remove_numbers(text)
  text = remove_spaces(text)
  text = sent_tok(text)
  return(text)
```

```python
def preprocess1(text):
  text = normalize(text)
  text = remove_stopwords(text)
  text = remove_numbers(text)
  text = stem_text(text)
  return(text)

```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

```

```python
def lemma_text(text):
  t  = ' '.join(lemmatizer.lemmatize(a) for a in word_tokenize(text))
  return t
```

```python
def preprocess3(text):
  text = normalize(text)
  text = remove_stopwords(text)
  text = remove_numbers(text)
  text = lemma_text(text)
  return(text)
```
